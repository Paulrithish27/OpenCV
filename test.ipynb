```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread("/content/sample_data/1.jpg")
cv2_imshow(img)
```

```python
img=cv2.imread("/content/sample_data/1.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```python
img=cv2.imread("/content/sample_data/1.jpg")
resized=cv2.resize(img,(200,200))
cv2_imshow(resized)
```

```python
img=cv2.imread("/content/sample_data/1.jpg")
add_circle=cv2.circle(img,(340,155),56,(255,0,0),6)
cv2_imshow(add_circle)
add_rect=cv2.rectangle(img,(80,50),(200,200),(0,255,0),3)
cv2_imshow(add_rect)
```

```python
img=cv2.imread("/content/sample_data/1.jpg")
rotation_matrix=cv2.getRotationMatrix2D((img.shape[1]/2,img.shape[0]/2),55,1)
rotated_image=cv2.warpAffine(img,rotation_matrix,(img.shape[1],img.shape[0]))
cv2_imshow(rotated_image)
```
